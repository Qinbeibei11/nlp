# 分词

参考文献: http://www.isnowfy.com/introduction-to-chinese-segmentation/

1. 词典匹配 & 2. 统计模型

基于字典的分词还是会暴露出很多的问题，最主要的问题就是歧义的问题，比如“武汉市长江大桥”，不同的分词可能会变成“武汉/市长/江大桥”和“武汉市/长江/大桥”，显然字典匹配是不能解决这样的歧义问题的，于是有了统计的分词算法



一元统计模型 $$ argmax \prod P(A_i) $$

n元统计模型 $$ argmax \prod P(A_i|A_{i-1},A_{i-2},...,A_{i-n}) $$

进阶:   
隐马尔可夫模型HMM（Hidden Markov Model）  
最大熵模型ME（Maximum Entropy）  
最大熵马尔可夫模型MEMM（Maximum-entropy Markov model）  
条件随机场CRF（Conditional Random Field）
