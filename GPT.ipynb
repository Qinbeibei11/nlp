```python
import gpt_2_simple as gpt2
import os
import requests
# import tensorflow as tf

model_name = "124M"
if not os.path.isdir(os.path.join("checkpoint", model_name)):
	print(f"Downloading {model_name} model...")
	gpt2.download_gpt2(model_name=model_name)   # model is saved into current directory under /models/124M/
```

# Directly load the downloaded model and generate some text.

```python
import gpt_2_simple as gpt2
import tensorflow as tf

tf.compat.v1.reset_default_graph()

sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name = '124M/') # The generated model checkpoints are by default in /checkpoint/run1. You can pass a run_name parameter to finetune and load_gpt2 if you want to store/load multiple models in a checkpoint folder.

gpt2.generate(sess)
```

```python
single_text = gpt2.generate(sess, return_as_list=True)[0]
print(single_text)
```
