```python
import pandas as pd

data = pd.read_excel('cata_4370.xls')
data.head()
```

```python
chemicals = data['不合格项目││检验结果││标准值'].values
foods = data['食品名称'].values

N = len(foods)

filtered_chemicals = []

for i in range(N):
    if '标签' in chemicals[i]:
        continue
        
    if '║' in chemicals[i]:
        filtered_chemicals.append(chemicals[i].split('║')[0])
    else:
        filtered_chemicals.append(chemicals[i])
```

```python
LIST1 = sorted(set(filtered_chemicals))
```

# 分析

```python
import pandas as pd

data = pd.read_excel('cata_4370.xls')
```

```python
links = {}

chemicals = data['不合格项目││检验结果││标准值'].values
foods = data['食品名称'].values

N = len(data)

for i in range(N):
    if '标签' in chemicals[i]:
        continue
        
    for c in chem:
        if c in chemicals[i]:
            if (foods[i], c) not in links:
                links[(foods[i], c)] = 1
            else:
                links[(foods[i], c)] += 1
```
